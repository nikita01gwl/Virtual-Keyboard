```python
#The method we use is that we split the screen into parts. If we blink the eye by pointing to the left of the 
#screen, the left part of keyboard pops up and then every letter changes color one by one, when we blink on a 
#letter, that letter is printed

import cv2
import numpy as np

#Black screen where we draw keys

keyboard = np.zeros((600,1000,3),np.uint8)

keys_set_1 = {0: "Q" , 1: "W", 2:"E", 3:"R", 4:"T",
             5: "A" , 6: "S", 7:"D", 8:"F", 9:"G",
             10: "Z" , 11: "X", 12:"C", 13:"V", 14:"B"
             }

def letter(letter_index,text,letter_light):
    
    #Draw keys on this keyboard
    #x = 0
    #y = 0
    
    #1st row
    if(letter_index==0):
        x=0
        y=0
        
    elif(letter_index==1):
        x=200
        y=0
        
    elif(letter_index==2):
        x=400
        y=0
        
    elif(letter_index==3):
        x=600
        y=0
        
    elif(letter_index==4):
        x=800
        y=0
        
    #2nd row
    if(letter_index==5):
        x=0
        y=200
        
    elif(letter_index==6):
        x=200
        y=200
        
    elif(letter_index==7):
        x=400
        y=200
        
    elif(letter_index==8):
        x=600
        y=200
        
    elif(letter_index==9):
        x=800
        y=200
        
    #3rd row
    if(letter_index==10):
        x=0
        y=400
        
    elif(letter_index==11):
        x=200
        y=400
        
    elif(letter_index==12):
        x=400
        y=400
        
    elif(letter_index==13):
        x=600
        y=400
        
    elif(letter_index==14):
        x=800
        y=400
        
        
    
        
        
    height = 200
    width = 200

    th = 3 #thickness

    #First letter A
    if letter_light is True:
        cv2.rectangle(keyboard, (x+th,y+th), (x+width-th,y+height-th), (255,255,255), -1)
    else:
        cv2.rectangle(keyboard, (x+th,y+th), (x+width-th,y+height-th), (255,0,0), th)


    #Put text inside rectangle and in the center
    font_letter = cv2.FONT_HERSHEY_PLAIN
    #text = "A"
    font_scale = 10
    font_th = 4

    #Get size of text. This is done to put the text in the middle of rectangle
    text_size = cv2.getTextSize(text,font_letter, font_scale,font_th)[0]
    width_text, height_text = text_size[0], text_size[1]

    text_x = int((width-width_text)/2) + x
    text_y = int((height+height_text)/2) + y
    
    cv2.putText(keyboard, text, (text_x,text_y), font_letter, font_scale, (255,0,0), font_th)


for i in range(15):
    letter(i,keys_set_1[i],True)
        
        
cv2.imshow("keyboard",keyboard)


cv2.waitKey(0)
cv2.destroyAllWindows()

_
```
